```python
#modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#line graph
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices in USD')

x= plt.plot(gas.Year, gas.USA, label="USA")
y= plt.plot(gas.Year, gas.Canada, label="Canada")
z =plt.plot(gas.Year, gas['South Korea'], label="South Korea")
plt.plot(gas.Year, gas.UK, label="UK")
plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("USDOLLARS")

plt.legend()
plt.show()
```

Fifa data



```python
fifa = pd.read_csv('fifa_data.csv')

fifa
```

Histogram

```python
bins=[30,40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)
plt.ylabel('number of players')
plt.xlabel("skill lvl")
plt.title("Player skills in FIFA 18")

plt.show()
```

PIE chart

```python
left= fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right= fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['left', 'right']
plt.pie([left, right], labels=labels, autopct='%.2f %%')

plt.title("Fifa footballer foot preferred")
plt.show()
```
