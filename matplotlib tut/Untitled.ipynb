```python
#modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#line graph
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices in USD')

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)
plt.plot(gas.Year, gas['South Korea'])

plt.xticks(gas.Year)

plt.legend()
plt.show()
```
